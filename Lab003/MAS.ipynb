```python
#función que verifica si un valor es negativo, en caso sea así, pide nuevamente el ingreso de datos
def verificarNegativo(x):   
   if (x<0):
    x1 = float(input('El valor ingresado no puede ser negativo, vuelva a ingresar otro valor : '))
    return x1
   else: 
    return x
```

```python
#función para graficar la posición respecto al tiempo
def graficarPosicion(ti, tf, A, w, phi):
  tiempo = []
  posicion = []
  t = ti
  while t < tf:
    tiempo.append(t)
    posicion.append(A * math.cos(math.radians(w*t + phi)))
    t = t + 0.5
  plt.ion()
  plt.plot(tiempo, posicion)
  plt.xlabel('Tiempo en segundos')
  plt.ylabel('Posición en m')
  plt.title('Cambio de posición en el tiempo')
  plt.grid(True)
  plt.draw()
  plt.pause(0.1)
  plt.ioff()
```

```python
#función para graficar la velocidad respecto al tiempo
def graficarVelocidad(ti, tf, A, w, phi):
  tiempo = []
  velocidad = []
  t = ti
  while t < tf:
    tiempo.append(t)
    velocidad.append(-A * w * math.sin(math.radians(w*t*phi)))
    t = t + 0.5
  plt.ion()
  plt.plot(tiempo, velocidad)
  plt.xlabel('Tiempo en segundos')
  plt.ylabel('Velocidad en m/s')
  plt.title('Cambio de velocidad en el tiempo')
  plt.grid(True)
  plt.draw()
  plt.pause(0.1)
  plt.ioff()
```

```python
# función para calcular y graficar la velocidad y aceleración máxima en un Movimiento Armónico Simple
def MAS():
  A = float (input('Ingrese la amplitud en metros : '))
  ti = float (input('Ingrese el tiempo inicial en segundos : '))
  ti = verificarNegativo(ti)
  tf = float (input('Ingrese el tiempo final en segundos : '))
  tf = verificarNegativo(tf)
  w = float (input('Ingrese la frecuencia  angular en rad/s : '))
  phi = float(input('Ingrese la fase en grados: '))
  
  velocidad_max = w * A
  aceleracion_max = pow(w, 2) * A
  print('La velocidad máxima es : ', round(velocidad_max, 2), ' m/s')
  print('La aceleración máxima es : ', round(aceleracion_max, 2), ' m/s²\n')
  graficarPosicion(ti, tf, A, w, phi)
  print()
  graficarVelocidad(ti, tf, A, w, phi)
```

```python
#main
print('----MOVIMIENTO ARMÓNICO SIMPLE---\n')
nota1()
continuar = 1
while continuar == 1:
  MAS()
  continuar = int(input('Ingrese 1 si desea continuar, caso contrario ingrese cualquier otro valor\n'))
```

### **2. Escriba la ecuación 4 en código computacioanl para resolver ejercicios de Oscilaciones Forzadas de forma generalizada.**


```python
#función para calcular y graficar la velocidad y aceleración máxima en un Movimiento Aramónico Simple
def MAS_Forzado():
  ti = float (input('Ingrese el tiempo inicial en segundos : '))
  ti = verificarNegativo(ti)
  tf = float (input('Ingrese el tiempo final en segundos : '))
  tf = verificarNegativo(tf)
  phi = float(input('Ingrese la fase en grados: '))
  w0 = float(input('Ingrese la frecuencia angular inicial en rad/s : '))
  w = float (input('Ingrese la frecuencia angular en rad/s : '))
  F0 = float (input('Ingrese la fuerza en Newton : '))
  m = float(input('Ingrese la masa en Kg : '))
  if m==0:
    print('No se puede calcular : ')
    m = float(input('Ingrese nuevamente la masa en Kg : '))
  else:
    A = (F0 / m) * (math.sqrt(pow(pow(w, 2) - (pow(w0,2)),2)) + pow((w/m), 2))

  velocidad_max = w * A
  aceleracion_max = pow(w, 2) * A
  print('La velocidad máxima es : ', round(velocidad_max, 2), 'm/s')
  print('La aceleración máxima es : ', round(aceleracion_max, 2), 'm/s²')
  graficarPosicion(ti, tf, A, w, phi)
  print()
  graficarVelocidad(ti, tf, A, w, phi)
```

```python
#main
print('----MOVIMIENTO ARMÓNICO SIMPLE FORZADO---\n')
nota1()
continuar = 1
while continuar == 1:
  MAS_Forzado()
  continuar = int(input('Ingrese 1 si desea continuar, caso contrario ingrese cualquier otro valor\n'))
```

### **3. Escriba las ecuaciones 6 y 8 en código computacional para resolver ejercicios de Oscilaciones Eléctricas de forma generalizada**

6 : Circuito LC   
8 : Circuito LCR

```python
#función para graficar la carga respecto al tiempo en LC
def graficarCargaLC(tf, Qi, L, C, phi):
  tiempo = []
  carga = []
  t = 0
  while t < tf:
    tiempo.append(t)
    w = 1 / math.sqrt(L * C)
    carga.append(Qi * math.sin(math.radians((w * t) + phi)))
    t = t + 0.1
  plt.ion()
  plt.plot(tiempo, carga)
  plt.xlabel('Tiempo en segundos')
  plt.ylabel('Carga en colombios ')
  plt.title('Cambio de la carga en el tiempo')
  plt.grid(True)
  plt.draw()
  plt.pause(0.1)
  plt.ioff()
```

```python
#función para carcular la carga en un Circuito LC
def circuitoLC():
  L = float(input('Ingrese la inductancia L en henrios: '))
  C = float(input('Ingrese la capacitancia C en faradios : '))
  Qi = float(input('Ingrese la Carga inicial en culombios: '))
  t = float(input('Ingrese el tiempo en segundos : '))
  t = verificarNegativo(t)
  phi = float(input('Ingrese la fase en grados : '))

  w = 1 / math.sqrt(L * C)
  Carga = Qi * math.sin(math.radians((w * t) + phi))
  print('La carga en el tiempo', t, 'segundos es:', round(Carga,2), ' µ')
  
  graficarCargaLC(t, Qi, L, C, phi)
```

```python
#función para graficar la carga respecto al tiempo en LCR
def graficarCargaLCR(tf, Qi, w, gamma, phi):
  tiempo = []
  carga = []
  t = 0
  while t < tf:
    tiempo.append(t)
    carga.append(Qi * math.exp(-gamma * t) * math.sin(math.radians(w * t + phi)))
    t = t + 0.1
  plt.ion()
  plt.plot(tiempo, carga)
  plt.xlabel('Tiempo en segundos')
  plt.ylabel('Carga en colombios ')
  plt.title('Cambio de la carga en el tiempo')
  plt.grid(True)
  plt.draw()
  plt.pause(0.1)
  plt.ioff()
```

```python
#función para carcular la carga en un Circuito LCR
def circuitoLCR():
  L = float(input('Ingrese la inductancia L en henrios: '))
  C = float(input('Ingrese la capacitancia C en faradios : '))
  R = float(input('Ingrese la resistencia R en ohmios : '))
  Qi = float(input('Ingrese la Carga inicial en culombios: '))
  t = float(input('Ingrese el tiempo en segundos : '))
  t = verificarNegativo(t)
  phi = float(input('Ingrese la fase en grados : '))

  gamma = R / (2 * L)
  w = 1 / math.sqrt(L * C)
  Carga = (Qi * math.exp(-gamma * t) * math.sin(math.radians(w * t + phi)))
  print('La carga en el tiempo', t, 'segundos es:', round(Carga,2), ' µ')
  
  graficarCargaLCR(t, Qi, w, gamma, phi)
```

```python
#menu
nota2()
continuar = 1
while continuar == 1:
  print('1: Circuito LC\n2: Circuito LCR\n')
  option = int(input('Elige una opción : '))
  if (option == 1):
    print('----CIRCUITO LC---\n')
    nota2()
    circuitoLC()
  if (option == 2):
    print('----CIRCUITO LCR---\n')
    nota2()
    circuitoLCR()
  else:
    print('Ingrese una opción correcta ')
  continuar = int(input('Ingrese 1 si desea continuar, caso contrario ingrese cualquier otro valor\n'))
```
