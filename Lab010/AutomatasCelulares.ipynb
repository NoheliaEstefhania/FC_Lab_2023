##### Matriz, de (15 x 30) con 30 iteraciones.


```python
'''
ti = tablero inicial
tn = tablero nuevo
'''
#dimensiones del tablero
filas = 15
columnas = 30

#crear tablero
ti = np.zeros((filas, columnas))
ti[0, 15] = 1

def update():
  tn = np.zeros((filas, columnas))
  for i in range (filas):
    for j in range (columnas):
      #regla de selección
      if(ti[i,(j-1+columnas)%columnas] == 0 and ti[i, j] == 0 and ti[i, (j+1)%columnas] == 0):
        tn[(i+1)%filas, j] = 0 #000 = 0
      elif(ti[i,(j-1+columnas)%columnas] == 0 and ti[i, j] == 0 and ti[i, (j+1)%columnas] == 1):
        tn[(i+1)%filas, j] = 1 #001 = 1
      elif(ti[i,(j-1+columnas)%columnas] == 0 and ti[i, j] == 1 and ti[i, (j+1)%columnas] == 0):
        tn[(i+1)%filas, j] = 1 #010 = 1
      elif(ti[i,(j-1+columnas)%columnas] == 0 and ti[i, j] == 1 and ti[i, (j+1)%columnas] == 1):
        tn[(i+1)%filas, j] = 1 #011 = 1
      elif(ti[i,(j-1+columnas)%columnas] == 1 and ti[i, j] == 0 and ti[i, (j+1)%columnas] == 0):
        tn[(i+1)%filas, j] = 1 #100 = 1
      elif(ti[i,(j-1+columnas)%columnas] == 1 and ti[i, j] == 0 and ti[i, (j+1)%columnas] == 1):
        tn[(i+1)%filas, j] = 0 #101 = 0
      elif(ti[i,(j-1+columnas)%columnas] == 1 and ti[i, j] == 1 and ti[i, (j+1)%columnas] == 0):
        tn[(i+1)%filas, j] = 0 #110 = 0
      else:
        tn[(i+1)%filas, j] = 0 #111 = 0
  #copiamos las celdas vivas de la matriz anterior a la matriz nueva
  tn[ti == 1] = 1

  return tn

#función para mostrar el tablero
def mostrar_tablero():
    for i in range(filas):
        for j in range(columnas):
            if ti[i, j] == 1:
                print('◼', end=' ')
            else:
                print('◻', end=' ')
        print()

#iteraciones -> tiempo
iteraciones = 30
for x in range(iteraciones):
    print('Iteración:', x+1)
    mostrar_tablero()
    ti = update()
    print()
```
