```python
#función que muestra algunos datos de la tierra
def datos():
  print(' ---- Datos----')
  print('masa de la tierra m = 5.972 * 10²⁴ Kg')
  print('radio r = 6371 km ')
  print('la constante gravitacional  G = 6.673 * 10^-11')
  print('la constante  K = 2.97 * 10 ^-19 s/m³')
  print('la distancia media desde el sol a la tierra es de 1.496 * 10 ^11 m\n')
```

```python
#función para determinar la gravedad de cualquier planeta en función a su radio y masa
def gravedadPlaneta(nombre, r, a, b):
  G = 6.673 * math.pow(10, -11) # constante gravitacional
  rMetros = r * 1000 # radio del planeta en metros
  masaPlaneta = a * math.pow(10, b) # masa de la tierra en m³
  gravedad = (G * (masaPlaneta))/(math.pow(rMetros, 2)) # en m/s²
  print('--> La gravedad del planeta ' , nombre , ' es ', round(gravedad, 2), ' m/s²')
```

### **2. Del problema anterior realice un código para poder determinar la densidad de cualquier planeta del sistema planetario solar.**

```python
#función para determinar la desidad de cualquier planeta en función a su radio y masa
def desidadPlaneta(nombre, r, a, b):
  G = 6.673 * math.pow(10, -11)
  rMetros = r * 1000 # radio del planeta en metros
  masaPlaneta = a * math.pow(10, b) # masa de la tierra
  volumen = (4/3) *  math.pi * math.pow(rMetros, 3) #volumen del planeta en Km^3
  densidad = masaPlaneta / volumen #densidad del planeta en kg/m3 = 5515
  print('--> La densidad del planeta ' , nombre , ' es ', round(densidad, 2), ' Kg/m³')
```

```python
#main
datos()
nombre = input('Ingrese el nombre del planeta: ')
r = float(input('Ingrese la distancia del planeta al sistema planetario solar en Km : '))
print('Ingrese la masa del planeta (a) * 10 a la (b)')
print('Por ejemplo si a = 5.972 y b = 24, entonces la masa del planeta es 5.972 * 10²⁴')
a = float(input('a = '))
b = float(input('b = '))
desidadPlaneta(nombre, r, a, b)
```

```python
#función que verifica si un valor es negativo, en caso sea así, pide nuevamente el ingreso de datos
def verificarNegativo(x):   
   if (x<0):
    x1 = float(input('El valor ingresado no puede ser negativo, vuelva a ingresar otro valor : '))
    return x1
   else: 
    return x
```

```python
#segunda ley de kepler, determinación del área en un tiempo t
def area_SegundaLey(r, a, b, t):
  Mp = a * math.pow(10, b) # masa de del planeta
  rMetros = r * 1000 # radio del planeta en metros
  velocidad = 4/3 *  math.pi * math.pow(rMetros, 3)
  # L: momento angular
  L = Mp * rMetros * velocidad  #L se puede calcular directamente como el producto de la masa del planeta por su velocidad y su distancia al centro del Sol
  area = (L / 2 * Mp ) * t
  print('El área barrida en el tiempo ', t, ' segundos, es ', round(area, 2), ' m³')
```
