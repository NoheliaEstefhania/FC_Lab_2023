```python
#función para determinar la gravedad de cualquier planeta en función a su radio y masa
def gravedadPlaneta(nombre, r, a, b):
  G = 6.673 * math.pow(10, -11) # constante gravitacional
  rMetros = r * 1000 # radio del planeta en metros
  masaPlaneta = a * math.pow(10, b) # masa de la tierra en m³
  gravedad = (G * (masaPlaneta))/(math.pow(rMetros, 2)) # en m/s²
  print('--> La gravedad del planeta ' , nombre , ' es ', round(gravedad, 2), ' m/s²')
```

### **2. Del problema anterior realice un código para poder determinar la densidad de cualquier planeta del sistema planetario solar.**

```python
#función para determinar la desidad de cualquier planeta en función a su radio y masa
def desidadPlaneta(nombre, r, a, b):
  G = 6.673 * math.pow(10, -11) # constante gravitacional
  rMetros = r * 1000 # radio del planeta en metros
  masaPlaneta = a * math.pow(10, b) # masa de la tierra
  volumen = (4/3) *  math.pi * math.pow(rMetros, 3) #volumen del planeta en Km³
  densidad = masaPlaneta / volumen #densidad del planeta en kg/m³ = 5515
  print('--> La densidad del planeta ' , nombre , ' es ', round(densidad, 2), ' Kg/m³')
```

```python
#main
datos()
nombre = input('Ingrese el nombre del planeta: ')
r = float(input('Ingrese el radio del planeta Km : '))
print('Ingrese la masa del planeta (a) * 10 a la (b)')
print('Por ejemplo si a = 5.972 y b = 24, entonces la masa del planeta es 5.972 * 10²⁴')
a = float(input('a = '))
b = float(input('b = '))
desidadPlaneta(nombre, r, a, b)
```

```python
#función que calcula la velocidad del planeta en m/s
def velocidadPlaneta(r):
  #r = 1.496 * math.pow(10, 11) # metros
  G = 6.673 * math.pow(10, -11) # constante gravitacional
  Ms = 1.989 * math.pow(10, 30) # Kg
  velocidad = math.sqrt((G * Ms)/ r)
  return velocidad
  #print('velocidad = ', velocidad, ' m/s')
```

```python
#segunda ley de kepler, determinación del área barrida en un tiempo t
def area_SegundaLey(r, Mp, t):
  #Mp = 5.972 * math.pow(10, 24)
  #r = 1.496 * math.pow(10, 11) # distancia desde el centro de la tierra al centro del sol en metros
  #t = 1
  G = 6.673 * math.pow(10, -11) # constante gravitacional
  velocidad = velocidadPlaneta(r)
  #L se puede calcular directamente como el producto de la masa del planeta por su velocidad y su distancia al centro del Sol
  L = Mp * r * velocidad # L: módulo del momento angular
  print('Momento angular de la tierra respecto al sol  L = ', round(L, 2))
  area = (L / 2 * Mp ) * t
  print('--> El área barrida en el tiempo ', t, ' segundos, es ', round(area, 2), ' m³')
```

### **4. Implementar un código computacional para determinar la solución de la tercera ley de Kepler para cualquier planeta que describa una órbita elíptica.**

```python
#imagen se cálculos previos
from IPython.display import Image
Image('/content/drive/MyDrive/FC_Images/terceraLeyKepler.png')
```

```python
#tercera ley de kepler
def terceraLey(nombre, r): # r: distancia media del sol al planeta
  K = 2.97 * math.pow(10, -19) # s/m³
  T = math.sqrt(K * math.pow(r, 3))# periodo orbital en segundos
  Tdias = T / 24 / 60 / 60 # horas, minutos y segundos
  print('--> El periodo orbital de ', nombre , ' es ', round(T, 2), ' segundos  = ', round(Tdias, 2), ' días' )
```

```python
#main
datos()
nombre = input('Ingrese el nombre del planeta: ')
print('Ingrese la distancia desde el centro del planeta al centro del sol en m, (a) * 10 a la (b)')
print('Por ejemplo si a = 1.496 y b = 11, entonces la masa del planeta es 1.989 * 10¹¹')
a = float(input('a = '))
b = float(input('b = '))
r = a * math.pow(10, b)
terceraLey(nombre, r)
```
